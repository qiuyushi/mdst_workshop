```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('states_edu.csv')
df.head()
```

```python
# want to focus on the math score of 8th grade
df["AVG_MATH_8_SCORE"].describe()
```

```python
df.count()
```

```python
df.dropna(subset=['AVG_MATH_8_SCORE'], inplace=True)
```

```python
df['ENROLL'] = df['ENROLL'].fillna(df.ENROLL.mean())
```

```python
df['TOTAL_REVENUE'] = df['TOTAL_REVENUE'].fillna(df.FEDERAL_REVENUE+df.STATE_REVENUE+df.LOCAL_REVENUE)
```

```python
df['TOTAL_REVENUE'].isna().sum()
```

```python
df['TOTAL_REVENUE'] = df['TOTAL_REVENUE'].fillna(df.TOTAL_REVENUE.mean())
```

```python
df['INSTRUCT_EXP_PERC'] = df.INSTRUCTION_EXPENDITURE / df.TOTAL_EXPENDITURE
```

```python
df.count()
```

```python
# plotting
df.AVG_MATH_8_SCORE.hist()
plt.xlabel('scores')
plt.ylabel('count')
plt.title('Math grade distribution')
```

```python
df.groupby('YEAR').AVG_MATH_8_SCORE.mean().plot()
plt.ylabel('score')
plt.title('Math grade reading score over time')
# math grades seems to increase then drop
```

```python
df.plot.scatter(x='ENROLL',y='AVG_MATH_8_SCORE')
plt.xlabel('8th grade enrollment')
plt.ylabel('8th grade reading score')
```

```python
df.plot.scatter(x='INSTRUCT_EXP_PERC',y='AVG_MATH_8_SCORE')
plt.xlabel('8th grade enrollment')
plt.ylabel('8th grade reading score')
## centers around .5
```

```python
df.plot.scatter(x='TOTAL_REVENUE',y='AVG_MATH_8_SCORE')
plt.xlabel('8th grade enrollment')
plt.ylabel('8th grade reading score')
## E = 1
```

```python
df.plot.scatter(x='AVG_MATH_4_SCORE',y='AVG_MATH_8_SCORE')
plt.xlabel('4th grade math score')
plt.ylabel('8th grade math score')
## clear linear relation between 4 and 8th grade
```

```python
df.plot.scatter(x='AVG_MATH_4_SCORE',y='AVG_MATH_8_SCORE')
plt.xlabel('8th grade enrollment')
plt.ylabel('8th grade reading score')
```
